```python
import pandas as pd

df = pd.read_csv('Netflix_Raw.csv')
```

```python
df.head(10)
```

```python
df.columns
```

```python
print("Missing Values:\n",df.isnull().sum())
```

```python
print("Duplicate Values:\n",df.duplicated().sum())
```

```python
print("Unique Values:\n",df.nunique())
```

```python
df['title'] = df['title'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')
df
```

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.strftime('%d-%m-%Y')
df
```

```python
df.drop_duplicates(inplace=True)
df
```

```python
df.columns = [col.title() for col in df.columns]
df
```

```python
df['Rating'] = df['Rating'].str.upper().replace({'PG 13': 'PG-13', 'TV-14': 'TV-14', 'TV-MA': 'TV-MA'})
df
```

```python
df['Date_Added'] = df['Date_Added'].fillna('Unknown')
df
```

```python
df.to_csv('Cleaned_Netflix.csv')
df
```
